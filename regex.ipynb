```python
names=["Alice", "Bob", "Charlie","David","Nick Franklin"]
regex= '^\w+ \w+$'
for n in names:
    result=re.search(regex,n)
    if result:
        print(n)
```

```python
text = "the price is 100 rs and i got 40 rs with me "

numbers = re.findall(r"\d+\.\d+|\d+",text)

print(numbers)
```
